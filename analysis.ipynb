```python
import matplotlib.pyplot as plt
# brute force vs sequential barnes hut
# N randomly generated points in a 1000 x 1000 box for 1 iteration
# theta = 0.5 =larger theta, faster and worst quality
N = [10, 100, 500, 1_000, 5_000, 10_000, 25_000, 50_000]
bf = [0, 0, 8, 20, 336, 1372, 8442, 34294]
bh = [0, 0, 4, 9, 56, 132, 418, 954]

N = [10, 100, 500, 1_000, 5_000, 10_000, 25_000, 50_000]
bf = [0, 0, 8, 20, 336, 1372, 8442, 34294]
bh = [0, 0, 4, 9, 56, 132, 418, 954]

plt.figure(figsize=(10, 6))
plt.plot(N, bf, 'o-', label='Brute Force O(N²)', color='tomato')
plt.plot(N, bh, 'o-', label='Barnes-Hut O(N log N)', color='teal')

plt.xlabel('Number of Bodies (N)')
plt.ylabel('Time per Iteration (ms)')
plt.title('Brute Force vs Barnes-Hut')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('speedup.png', dpi=150)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cpp_benchmark_results.csv")

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))

for theta, group in df.groupby("theta"):
    group = group.sort_values("N")
    ax1.plot(group["N"], group["speedup"], marker="o", label=f"θ={theta}")
    ax2.plot(group["N"], group["avg_rel_error_pct"], marker="o", label=f"θ={theta}")
    ax3.plot(group["N"], group["barnes_hut_ms"], marker="o", label=f"BH θ={theta}")

# plot brute force once (same across thetas)
bf = df.drop_duplicates(subset="N").sort_values("N")
ax3.plot(bf["N"], bf["brute_force_ms"], marker="s", color="black", linestyle="--", label="Brute Force")

ax1.set_xscale("log")
ax1.set_xlabel("Number of Bodies")
ax1.set_ylabel("Speedup (Brute Force / Barnes-Hut)")
ax1.set_title("Barnes-Hut Speedup vs Brute Force")
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.set_xscale("log")
ax2.set_xlabel("Number of Bodies")
ax2.set_ylabel("Average Relative Force Error (%)")
ax2.set_title("Barnes-Hut Approximation Error")
ax2.legend()
ax2.grid(True, alpha=0.3)

ax3.set_xscale("log")
ax3.set_yscale("log")
ax3.set_xlabel("Number of Bodies")
ax3.set_ylabel("Runtime (ms)")
ax3.set_title("Raw Runtime")
ax3.legend()
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("benchmark_plots.png", dpi=150)
plt.show()
```
